```python
import numpy as np
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
```

```python
dataset = pd.read_csv('Amazon dataset.csv')
```

```python
# Define the mapping of current column names to shorter names
column_name_mapping = {
'id': 'ProdID',
'name': 'ProdName',
'categories': 'Category',
'imageURLs': 'ImageURL',
'reviews.didPurchase':'didPurchase',
'reviews.rating': 'Rating',
'reviews.username': 'Username',
'Product Image Url': 'ImageURL',
'reviews.date':'Date'
}
# Rename the columns using the mapping
dataset.rename(columns=column_name_mapping, inplace=True)
```

```python
dataset
```

# 1 step : Encode the Transactions

```python
transactions = dataset.groupby(['Date','Username'])['ProdName'].apply(list).tolist()
```

```python
# Initialize the TransactionEncoder
encoder = TransactionEncoder()
```

```python
encoded_transactions = encoder.fit_transform(transactions)
```

```python
basket = pd.DataFrame(encoded_transactions, columns=encoder.columns_)
```

```python
basket
```

```python
basket = basket.astype(int)
```

```python
basket
```

```python
basket.to_csv("Transaction_encoded.csv",index=False)
```

```python
basket['Month'] = transactions['Month']
```

# 4 step : Get the monthly support

```python
monthly_support = basket.groupby('Month').mean() # get monthly support
```

# 5 step : Contextual Filtering Function

```python

def recommend_items_month(selected_month):
    # Set the number of top items to recommend
    top_n = 3
    
    # Convert the selected month string to a datetime object for manipulation
    selected_month_dt = pd.to_datetime(selected_month)
    
    # Initialize previous_year_month as one year before the selected month
    previous_year_month = (selected_month_dt - pd.DateOffset(years=1)).strftime('%Y-%m')

    # Loop to find a valid previous year month in the monthly_support DataFrame
    while previous_year_month not in monthly_support.index:
        # Subtract one year from the selected month
        selected_month_dt -= pd.DateOffset(years=1) 
        
        # Update previous_year_month to the new year-month string
        previous_year_month = selected_month_dt.strftime('%Y-%m') 
        
        # Print the current previous year month being checked
        #print(previous_year_month)

        # Check if the year has dropped below a certain limit to avoid infinite loops
        if selected_month_dt.year < 2000: 
            print("No data available for the previous years.")
            return  # Exit the function if no valid month is found

    # Extract the support values for the found previous year's selected month
    answer = monthly_support.loc[previous_year_month]

    # Rank the items by their support values in descending order
    ranked_items = answer.sort_values(ascending=False)

    # Print the top-N item names for the previous year
    for item in ranked_items.head(top_n).index:
        print(item)
```

```python
# Get the current date
current_date = pd.to_datetime("now")

# Format it as 'YYYY-MM'
formatted_date = current_date.strftime('%Y-%m')

recommend_items_month(formatted_date)
```
