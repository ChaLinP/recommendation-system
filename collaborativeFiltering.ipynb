```python
dataset = pd.read_csv('Amazon dataset.csv')
```

```python
dataset.head()
```

```python
dataset = dataset[['id','name','categories','imageURLs','reviews.didPurchase','reviews.rating','reviews.username','reviews.date']]
```

```python
dataset.head()
```

```python
# Define the mapping of current column names to shorter names

column_name_mapping = {

'id': 'ProdID',

'name': 'ProdName',

'categories': 'Category',

'imageURLs': 'ImageURL',

'reviews.didPurchase':'didPurchase',

'reviews.rating': 'Rating',

'reviews.username': 'Username',

'imageURLs': 'ImageURL',

'reviews.date' : 'Date',

}

# Rename the columns using the mapping

dataset.rename(columns=column_name_mapping, inplace=True)
 
```

```python
transactions = dataset.groupby(['Date','Username'])['ProdName'].apply(list).reset_index()
```

```python
transactions.head()
```

```python
transactions = dataset.groupby(['Date','Username'])['ProdName'].apply(list).tolist()
```

```python
from mlxtend.preprocessing import TransactionEncoder
```

```python
# Initialize the TransactionEncoder
encoder = TransactionEncoder()
```

```python
encoded_transactions = encoder.fit_transform(transactions)
```

```python
basket = pd.DataFrame(encoded_transactions, columns=encoder.columns_)
```

```python
basket.head()
```

```python
from mlxtend.frequent_patterns import apriori, association_rules
 
# Step 1: Apply the Apriori algorithm
# Find frequent itemsets with a minimum support of 0.2 (this can be adjusted)
frequent_itemsets = apriori(basket, min_support=0.01, use_colnames=True)
 
print("\nFrequent Itemsets:")
frequent_itemsets
```

```python
# Step 2: Generate association rules
# Set the minimum confidence level to 0.6 (this can be adjusted)
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.1)
 
print("\nAssociation Rules:")
rules
```

```python
from mlxtend.frequent_patterns import apriori, association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```python
# Function to get the top N most frequently bought products
def get_top_products(dataset, top_n=5):
    return dataset['ProdName'].value_counts().head(top_n).index.tolist()

# Function to recommend products based on username (without rules as a parameter)
def recommend_products_by_user(username, dataset, top_products, top_n=5):
 
    
    # Filter dataset for the specific user
    user_data = dataset[dataset['Username'] == username]
    
    # If no data is found for the user
    if user_data.empty:
        return "No data available for this user."
    
    # Get the list of products purchased by the user
    user_purchases = user_data['ProdName'].tolist()
    
    # Check if there are any recommendations based on previous purchases
    recommendations = []  # This will hold the recommendations based on user history

    # Ensure that rules are a DataFrame with the correct structure
    for _, rule in rules.iterrows():  # Assuming rules is a DataFrame
        antecedent = set(rule['antecedents'])  # Access 'antecedents' column
        if antecedent.issubset(set(user_purchases)):
            recommendations.extend(rule['consequents'])  # Access 'consequents' column
    
    # Remove duplicates
    recommendations = list(set(recommendations))
    
    # If no recommendations found, fall back to top popular products
    if not recommendations:
        recommendations = top_products

    # Return the top N recommendations
    return recommendations[:top_n]

# Example usage
top_products = get_top_products(dataset)  # Already generated
username = 'BySharon Lambert'  # Sample username
recommended_products = recommend_products_by_user(username, dataset, top_products)
print("Recommended Products:")
print(recommended_products)
```
