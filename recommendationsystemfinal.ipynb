```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords

# Download required resources
nltk.download('punkt')
nltk.download('stopwords')

# Define stop words
stop_words = set(stopwords.words('english'))

def process_text(text):
    text = text.casefold()  # Lowercase the text using casefold (more aggressive than lower())
    tokens = word_tokenize(text)  # Tokenize the text into words
    cleaned_tokens = [word for word in tokens if word.isalnum() and word not in stop_words]  # Remove stop words
    return cleaned_tokens

# Apply the function to the 'ProdName' column
product['ProdName'] = product['ProdName'].apply(lambda x: process_text(x))
product['Category'] = product['Category'].apply(lambda x: process_text(x))
```

```python
# Join list elements into strings for 'ProdName' and 'Category'
product['ProdName'] = product['ProdName'].apply(lambda x: ' '.join(x))  # Joining list elements in ProdName
product['Category'] = product['Category'].apply(lambda x: ' '.join(x))  # Joining list elements in Category

# Convert necessary columns to strings
product['ProdID'] = product['ProdID'].astype(str)
product['Rating'] = product['Rating'].astype(str)

# Create the 'Tags' column by concatenating the relevant columns
product['Tags'] = product['ProdID'] + ' ' + product['ProdName'] + ' ' + product['Category'] + ' ' + product['Rating']
```

```python
# Function to recommend similar products

def similar_products(product_name, product_df, cosinesimilarity_matrix, top_n=5):
    # Check if the product exists in the dataframe
    try:
        product_index = product_df[product_df['ProdName'].str.contains(product_name, case=False, na=False)].index[0]
    except IndexError:
        print(f"No product found matching '{product_name}' in the dataset.")
        return

    # Get cosine similarity scores for the selected product
    distances = cosinesimilarity_matrix[product_index]

    # Get most similar products excluding the input product itself
    products_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:top_n+1]

    # Print the top recommended products
    print(f"\nProducts similar to '{product_name}':")
    for i in products_list:
        print(product_df.iloc[i[0]]['ProdName'])

# Transform 'Tags' into a TF-IDF matrix
tfidf_matrix = vectorizer.fit_transform(product['Tags']).toarray()

# Compute cosine similarity matrix
cosinesimilarity_matrix = cosine_similarity(tfidf_matrix)
```

```python
# Calculate item-item similarity (ProdID)
item_similarity = cosine_similarity(user_item_matrix.T)  # Transpose the matrix to calculate item similarity
item_similarity_df = pd.DataFrame(item_similarity, index=user_item_matrix.columns, columns=user_item_matrix.columns)
```

```python
# Calculate user-user similarity (Username)
user_similarity = cosine_similarity(user_item_matrix)
user_similarity_df = pd.DataFrame(user_similarity, index=user_item_matrix.index, columns=user_item_matrix.index)
```

```python
# Create a mapping of ProdID to ProdName
id_to_name = dataset[['ProdID', 'ProdName']].drop_duplicates().set_index('ProdID')['ProdName'].to_dict()
```

```python
# Function to recommend items based on item-item similarity
def recommend_items_item_based(user, top_n=5):
    user_ratings = user_item_matrix.loc[user]
    similar_items = item_similarity_df.dot(user_ratings).sort_values(ascending=False)
    recommended_items = similar_items[~user_ratings.index.isin(user_ratings[user_ratings > 0].index)].head(top_n)
    return [id_to_name[item] for item in recommended_items.index]
```

```python
# Function to recommend items based on user-user similarity
def recommend_items_user_based(user, top_n=5):
    similar_users = user_similarity_df[user].sort_values(ascending=False).index
    similar_users_ratings = user_item_matrix.loc[similar_users]
    user_recommendations = similar_users_ratings.mean(axis=0).sort_values(ascending=False)
    recommended_items = user_recommendations[~user_item_matrix.loc[user].index.isin(user_item_matrix.loc[user][user_item_matrix.loc[user] > 0].index)].head(top_n)
    return [id_to_name[item] for item in recommended_items.index]
```

```python
user = 'Bill Clinton'  # Replace with an actual username from your dataset
item_based_recommendations = recommend_items_item_based(user)
user_based_recommendations = recommend_items_user_based(user)

def print_recommendations(title, recommendations):
    print(f"\n{title}:")
    for idx, product in enumerate(recommendations, start=1):
        print(f"{idx}. {product}")

print_recommendations(f"Item-based recommendations for {user}", item_based_recommendations)
print_recommendations(f"User-based recommendations for {user}", user_based_recommendations)
```

```python
def hybrid_recommendation(user, product_name, w_user_user=0.4, w_item_item=0.3, w_content=0.3):
    # Get recommendations from each filter
    user_user_score = recommend_items_user_based(user)
    item_item_score = recommend_items_item_based(user)

    # Call similar products and handle potential None return
    content_score = similar_products(product_name, product, cosinesimilarity_matrix, top_n=5)
    if content_score is None:
        # Handle the case where similar_products returns None (e.g., log error, return empty list)
        content_score = []  # or raise an exception, or handle differently based on your needs

    # Convert lists to sets for set operations
    user_user_set = set(user_user_score)
    item_item_set = set(item_item_score)
    content_set = set(content_score)

    # Combine recommendations using weighted hybrid approach
    final_recommendations = []
    for item in user_user_set.union(item_item_set).union(content_set):
        score = 0
        if item in user_user_set:
            score += w_user_user
        if item in item_item_set:
            score += w_item_item
        if item in content_set:
            score += w_content
        final_recommendations.append((item, score))

    # Sort recommendations by score and return top recommendations
    final_recommendations.sort(key=lambda x: x[1], reverse=True)
    return [item for item, score in final_recommendations]

# Example usage
user = 'Bill Clinton'
product_name = 'Laptop'
recommendations = hybrid_recommendation(user, product_name)
print(f'Hybrid recommendations for user {user} and item {product_name}: {recommendations}')
```

```python
# Function to calculate precision and recall
def precision_recall(user_recommended_items, user_relevant_items):
    # Convert lists to sets for easier comparison
    recommended_set = set(user_recommended_items)
    relevant_set = set(user_relevant_items)

    # Calculate true positives (relevant recommended items)
    true_positives = len(recommended_set.intersection(relevant_set))

    # Calculate precision and recall
    precision = true_positives / len(recommended_set) if len(recommended_set) > 0 else 0
    recall = true_positives / len(relevant_set) if len(relevant_set) > 0 else 0

    return precision, recall

# Example usage
user = 'Bill Clinton'  # The user for whom recommendations are being made

# Define relevant items (the actual items the user interacted with, e.g., highly rated or purchased)
user_relevant_items = dataset[dataset['Username'] == user]['ProdName'].unique().tolist()

# Get the recommended items from the hybrid recommendation system
recommended_items = hybrid_recommendation(user, 'Laptop')

# Calculate precision and recall
precision, recall = precision_recall(recommended_items, user_relevant_items)

print(f"Precision: {precision}, Recall: {recall}")
```

```python
import numpy as np

# Function to find the best weights for hybrid recommendation
def find_best_weights(user, product_name, step_size=0.1):
    best_precision = 0
    best_recall = 0
    best_weights = (0, 0, 0)  # Initialize best weights to zero

    # Iterate over weight combinations in steps of step_size
    for w_user_user in np.arange(0, 1 + step_size, step_size):
        for w_item_item in np.arange(0, 1 + step_size, step_size):
            for w_content in np.arange(0, 1 + step_size, step_size):
                # Ensure weights sum to 1 (or skip combinations that do not sum to 1)
                if abs(w_user_user + w_item_item + w_content - 1) < 1e-6:
                    # Get hybrid recommendations with current weights
                    recommendations = hybrid_recommendation(user, product_name, w_user_user, w_item_item, w_content)

                    # Calculate precision and recall
                    precision, recall = precision_recall(recommendations, user_relevant_items)

                    # Check if current weights give a better precision or recall
                    if precision > best_precision or (precision == best_precision and recall > best_recall):
                        best_precision = precision
                        best_recall = recall
                        best_weights = (w_user_user, w_item_item, w_content)
                        print(f"New best weights found: User-User: {w_user_user}, Item-Item: {w_item_item}, Content: {w_content}")
                        print(f"Precision: {precision}, Recall: {recall}")

    return best_weights, best_precision, best_recall

# Example usage
user = 'Bill Clinton'
product_name = 'Laptop'

# Define relevant items (the actual items the user interacted with, e.g., highly rated or purchased)
user_relevant_items = dataset[dataset['Username'] == user]['ProdName'].unique().tolist()

best_weights, best_precision, best_recall = find_best_weights(user, product_name)
print(f"Best weights: {best_weights}")
print(f"Best Precision: {best_precision}, Best Recall: {best_recall}")
```
